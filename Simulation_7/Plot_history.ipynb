### 2.1 - Load the relevant csv files

```python
np_dictionary = {}

dir_name = "output"
os_listdir = os.listdir(dir_name)
for file_name in os_listdir:
    file_path = os.path.join(dir_name, file_name)
    data =  pd.read_csv(file_path, delimiter=',', dtype=None, header=None)
    np_dictionary[file_name[:-4]] = data
```

<a name='2-3'></a>
### 2.2 - Plot the PDE in one dimension: fix x=-1 and varies t within [0, 2]

```python
impose_ylim = True
y_lim = [0,0.5]

real_values = np.asarray(np_dictionary["analytic"])[1,:]
real_values = np.reshape(real_values, (301,))[0:191]
t_values = np.arange(0, 1.91, 0.01)
# estimated_values_real = np.asarray(np_dictionary["monte_carlo"])[0,:]
# estimated_values_real = np.reshape(estimated_values_real, (31,))[0:19]
# t_estimated = np.arange(0, 1.9, 0.1)
estimated_values_imaginary_1 = np.asarray(np_dictionary["monte_carlo"])[1,:]
estimated_values_imaginary_1 = np.reshape(estimated_values_imaginary_1, (31,))[0:19]
t_estimated = np.arange(0, 1.9, 0.1)
estimated_values_imaginary_2 = - np.asarray(np_dictionary["monte_carlo_2"])[1,:]
estimated_values_imaginary_2 = np.reshape(estimated_values_imaginary_2, (31,))[0:19]
t_estimated = np.arange(0, 1.9, 0.1)

plt.figure(figsize=(12, 8))
plt.plot(t_values, real_values, label='Real values of u(t,-1)')
# plt.plot(t_estimated, estimated_values_real, label='Estimated values - real part')
plt.plot(t_estimated, estimated_values_imaginary_1, 'ro-', label='Estimated values 1 - imaginary part')
plt.plot(t_estimated, estimated_values_imaginary_2, 'go-', label='Estimated values 2 - negative of the imaginary part')

plt.ylabel('real vs estimated values')
if impose_ylim:
    plt.ylim(y_lim)
plt.xlabel('t values')
# plt.axvline(x=0.32,color='red')

plt.title('Real vs estimated values of u(t,-1)')
plt.legend(loc='upper right')
plt.show()
```
